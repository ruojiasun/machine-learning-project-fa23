```python
import pandas as pd
```

```python
# creating dataframes as df1 and df2 
df = pd.read_csv('31115647.csv')
df_sel_columns = df[['qn1','qns3','qns9b', 'gender','race','fpl']]

df_sel_columns = df_sel_columns.rename(columns={'qn1': 'recovery', 'qns3': 'home damage', 'qns9b': 'reduced hours', 'fpl':'income'})
for col in df_sel_columns.columns:
    print(df_sel_columns[col].value_counts())
    print()
```

```python
# map the recovery responses to 2 classes and balance
d_recovery = {'Largely back to normal': 'yes', 'Almost back to normal':'yes', 'Still somewhat disrupted':'no', 'Still very disrupted': 'no'}
df_cleaned = df_sel_columns
df_cleaned['recovery'] = df_cleaned['recovery'].map(d_recovery)

print(df_cleaned['recovery'].value_counts())

no_subset = df_cleaned.loc[df_cleaned['recovery'] == 'no', :]
num_no = round(len(no_subset)*1.1)
# print(num_no)

yes_subset = df_cleaned.loc[df_cleaned['recovery'] == 'yes', :]
sampled_yes = yes_subset.sample(num_no)

df_cleaned = pd.concat([no_subset, sampled_yes], ignore_index=True)

print(df_cleaned['recovery'].value_counts())

```

```python
# dropped don't know/refused for any feature
df_cleaned = df_cleaned[df_cleaned['home damage'] != "Don't know"]
df_cleaned = df_cleaned[df_cleaned['reduced hours'] != "Don't know"]
df_cleaned = df_cleaned[df_cleaned['income'] != "DK/RF"]

# balance race
race_list =['Black or African-American','White']
df_cleaned = df_cleaned[df_cleaned.race.isin(race_list)]


black_subset = df_cleaned.loc[df_cleaned['race'] == 'Black or African-American', :]
num_black = round(len(black_subset)*1.1)
# print(num_black)

white_subset = df_cleaned.loc[df_cleaned['race'] == 'White', :]
sampled_white = white_subset.sample(num_black)

df_cleaned = pd.concat([black_subset, sampled_white], ignore_index=True)

for col in df_cleaned.columns:
    print(df_cleaned[col].value_counts())
    print()

for col in df_cleaned.columns:
    print(df_cleaned[col].value_counts())
    print()
```

```python
df_cleaned = df_cleaned.sample(frac=1).reset_index(drop=True)

df_cleaned.to_csv('df_cleaned_2.csv', sep=',', index=False, encoding='utf-8')
```
